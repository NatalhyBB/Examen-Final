```python
# Parte 1:Fundamentos Numpy
import numpy as np

# 1. Arreglo del 1 al 20
arr = np.arange(1, 21)
print("1) Arreglo 1D (1 al 20):\n", arr, "\n")
print("═" * 60)

# 2. Selección de múltiplos de 3
multiplos_3 = arr[arr % 3 == 0]
print("2) Múltiplos de 3:\n", multiplos_3, "\n")
print("═" * 60)

# 3. Matriz aleatoria 5x5
np.random.seed(42)
matriz = np.random.rand(5, 5)
print("3) Matriz aleatoria 5x5:\n", matriz, "\n")
print("═" * 60)

# 4. Estadísticos básicos
media = np.mean(matriz)
desviacion = np.std(matriz)
maximo = np.max(matriz)

print("4) Estadísticos de la matriz:")
print(f"   → Media              : {media:.2f}")
print(f"   → Desviación estándar: {desviacion:.2f}")
print(f"   → Valor máximo       : {maximo:.2f}\n")
print("═" * 60)

# 5. Transformación a vector 1D
vector = matriz.flatten()
print("5) Vector 1D (matriz a plano):\n", vector, "\n")
```

```python
# Parte 2: Fundamentos Pandas

import pandas as pd
from tabulate import tabulate

# Datos de estudiantes
data = {
    "Nombre": ["Ana", "Luis", "María", "Pedro", "Sofía"],
    "Edad": [20, 21, 19, 22, 20],
    "Nota_Matemáticas": [18, 15, 16, 12, 19],
    "Nota_IA": [17, 14, 18, 13, 20]
}

df = pd.DataFrame(data)
col_alignment = ["center"] * len(df.columns)

print("1) Calificaciones de los estudiantes:")
print(tabulate(df, headers="keys", tablefmt="fancy_grid", showindex=False,  colalign=col_alignment))
print(" ")

# Promedio de cada columna numérica
promedios = df.mean(numeric_only=True).round(2)
promedios_df = promedios.reset_index()
promedios_df.columns = ["Columna", "Promedio"]

print("2) Promedio de cada columna numérica:")
print(tabulate( promedios_df, headers="keys", tablefmt="fancy_grid", showindex=False, colalign=["center", "center"]))
print(" ")

# Filtrar estudiantes con Nota_IA > 15
filtro_ia = df[df["Nota_IA"] > 15]
print("3) Estudiantes con Nota de IA mayor a 15")
print(tabulate( filtro_ia, headers="keys", tablefmt="fancy_grid", showindex=False, colalign=["center"] * len(filtro_ia.columns)))
print(" ")

# Agregar columna Promedio
df["Promedio"] = df[["Nota_Matemáticas", "Nota_IA"]].mean(axis=1).round(2)
print("4) DataFrame con columna 'Promedio'")
print(tabulate( df, headers="keys", tablefmt="fancy_grid", showindex=False, colalign=["center"] * len(df.columns)))
print(" ")

# Ordenar por 'Promedio' de mayor a menor
df_ordenado = df.sort_values(by="Promedio", ascending=False)
print("5) DataFrame ordenado por Promedio (descendente)")
print(tabulate( df_ordenado, headers="keys", tablefmt="fancy_grid", showindex=False, colalign=["center"] * len(df.columns)))
print(" ")
```

```python
# Parte 3: Algoritmia aplicada
# 1) Función que recibe lista de números
import numpy as np
import pandas as pd
from tabulate import tabulate

def analizar_lista(numeros):
    mayor = max(numeros)
    menor = min(numeros)
    promedio = sum(numeros) / len(numeros)
    return mayor, menor, promedio

entrada = input("Ingrese números separados por comas: ")
lista = [float(x) for x in entrada.split(",")]
mayor, menor, promedio = analizar_lista(lista)

print("\nResultados:")
print(f"➤ Mayor: {mayor}")
print(f"➤ Menor: {menor}")
print(f"➤ Promedio: {promedio:.2f}\n")

# 2) Función de activación Sigmoide con NumPy

def sigmoide(x):
    return 1 / (1 + np.exp(-x))
x = np.array([-2, -1, 0, 1, 2])
resultado = sigmoide(x)

print("2) Función Sigmoide aplicada a el arreglo: [-2, -1, 0, 1, 2]:")
print(resultado)
print("\n")

# 3) Clasificación de estudiantes

data = {
    "Nombre": ["Ana", "Luis", "María", "Pedro", "Sofía"],
    "Edad": [20, 21, 19, 22, 20],
    "Nota_Matemáticas": [18, 15, 16, 12, 19],
    "Nota_IA": [17, 14, 18, 13, 20]
}
df = pd.DataFrame(data)


df["Promedio"] = df[["Nota_Matemáticas", "Nota_IA"]].mean(axis=1).round(2)

def clasificar(promedio):
    return "Aprobado" if promedio >= 14 else "Desaprobado"

df["Estado"] = df["Promedio"].apply(clasificar)

print("3) Clasificación de estudiantes:")
print(tabulate(
    df,
    headers="keys",
    tablefmt="fancy_grid",
    showindex=False,
    colalign=["center"] * len(df.columns)
))
```

```python
# Parte 4: Mini-proyecto

import pandas as pd

datos = {
    'Producto': ['A', 'B', 'C', 'D', 'E'],
    'Ventas_2024': [120, 340, 560, 230, 150],
    'Ventas_2025': [200, 400, 600, 300, 180]
}

df = pd.DataFrame(datos)

df.to_excel("ventas.xlsx", index=False)
print("Archivo 'ventas.xlsx' creado")
print("")
df = pd.read_excel("ventas.xlsx")

# 1) Agregar columna de Crecimiento

df["Crecimiento"] = ((df["Ventas_2025"] - df["Ventas_2024"]) / df["Ventas_2024"]) * 100
df["Crecimiento"] = df["Crecimiento"].round(2)

tabla = tabulate(df, headers="keys", tablefmt="fancy_grid", showindex=False, colalign=("center", "center", "center", "center"))
print("1) Agregar columna de Crecimiento ")
print(tabla)

# 2) Identificar producto con mayor crecimiento

mayor_crecimiento = df.loc[df["Crecimiento"].idxmax()]

print("")
print("2) Identificar producto con mayor crecimiento")
print(tabla)

print("\nProducto con mayor crecimiento:")
print(mayor_crecimiento["Producto"], "→", mayor_crecimiento["Crecimiento"], "%")

```
